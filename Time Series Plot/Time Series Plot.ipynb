```python
df = pd.read_csv('Time_Series_datasets/daily-temperature.csv', parse_dates=True)
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
df.set_index('Date', inplace=True)
```

```python
df.head()
```

### Line and Scatter Plot

```python
df
```

```python
df.plot(color = 'r')
plt.ylabel('Temp')
plt.title('Line Plot')
plt.show()
```

```python
df.plot(style = '.', color = 'r')
plt.ylabel('Temp')
plt.title('Line Plot')
plt.show()
```

### SubPlots

```python
df
```

```python
#grouping the data together by groupby()
groups = df.groupby(pd.Grouper(freq = 'A'))
```

```python
keys = groups.groups.keys()
```

```python
keys
```

```python
groups.get_group('1982-12-31')
```

```python
for key in keys:
    print(key)
```

```python
key
```

```python
groups.get_group(key)['Temp'].values
```

```python
years = pd.DataFrame()
for key in keys: 
    years[key] = groups.get_group(key)['Temp'].values 
```

```python
years
```

```python
# Plotting it into 10 different subplots
years.plot(subplots = True, figsize = (10, 30))
plt.show()
plt.tight_layout()
```
