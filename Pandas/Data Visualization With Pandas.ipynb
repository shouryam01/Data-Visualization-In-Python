### Scatter Matrix and SubPlots

```python
from pandas.plotting import scatter_matrix
```

```python
df
```

```python
scatter_matrix(df, figsize = (8, 8), diagonal = 'kde', color = 'r')
plt.show()
```

KDE Plots - KDE stands for Kernel Density Estimation Plots

##### Plotting the subPlots

```python
df.plot()
```

```python
df.plot(subplots = True)
```

```python
# not sharing x axes
df.plot(subplots = True, sharex = False)
```

```python
#avoiding overlapping 
df.plot(subplots = True, sharex = False)
plt.tight_layout()
plt.show()
```

```python
# manage the subplots in 2 by 2 subplots
df.plot(subplots = True, sharex = False, layout = (2,2), figsize = (7, 7))
plt.tight_layout()
plt.show()
```
