### LinePlot

```python
df = pd.DataFrame(np.random.randn(100, 3), columns=['A', 'B', 'C'])
```

```python
df.head()
```

```python
df = df.cumsum()
```

```python
df.head()
```

```python
df.shape
```

```python
# plotting it using plotly
df.iplot()
```

```python
df.plot()
```

### Scatter Plot

```python
df.head()
```

```python
df.iplot(x = 'A', y = 'B', mode = 'markers', size = 20, title = 'Scatter Plot')
```

### Bar Plot

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
# using bar pot with plotly
```

```python
titanic.iplot(kind = 'bar', x = 'sex', y = 'survived', title = 'Titanic')
```

```python
titanic['sex'].value_counts()
```

```python
# plotting bar plot in this continous data
df.iplot(kind = 'bar', barmode = 'stack', bargap = 0.5)
```

```python
# Plotting it in horizontal axes
df.iplot(kind = 'barh', barmode = 'stack', bargap = 0.5)
```

### Box Plot and Area Plot

```python
df.head()
```

```python
df.iplot(kind= 'box')
```

### Area Plot

```python
df.iplot(kind = 'area', fill = True)
```

### 3D Plot

```python
# Plot in 3-Dimension using plotly and cufflinks
df.iplot(kind = 'surface')
```

```python
# changing colorscale
df.iplot(kind = 'surface', colorscale = 'rdylbu')
```

```python
cf.datagen.sinwave(10, 0.25).iplot(kind = 'surface')
```

### Spread Plot and Hist Plot

```python
df.iplot(kind = 'spread')
```

```python
df.iplot(kind = 'hist', bins = 25, barmode= 'overlay', bargap = 0.5)
```

### Bubble Plot and Heatmap

```python
bb = cf.datagen.bubble3d(5, 4, mode = 'stocks')
```

```python
bb.head()
```

```python
bb.iplot(kind = 'bubble3d', x = 'x', y = 'y', z = 'z', size = 'size')
```

### Heatmap

```python
df.corr()
```

```python
df.corr().iplot(kind = 'heatmap', colorscale = 'spectral')
```

```python
cf.datagen.heatmap(20, 20)
```

```python
cf.datagen.heatmap(20, 20).iplot(kind = 'heatmap', colorscale = 'spectral')
```
